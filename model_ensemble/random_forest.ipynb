### Explicación de Random Forest

**Random Forest** es un algoritmo de aprendizaje automático versátil y popular que se utiliza tanto para **clasificación** como para **regresión**. Pertenece a la familia de los algoritmos basados en árboles y se considera un método de *ensemble* (conjunto).

**¿Cómo funciona?**

1.  **Construcción de múltiples árboles:** En lugar de construir un solo árbol de decisión, Random Forest construye un gran número de árboles de decisión independientes durante la fase de entrenamiento.
2.  **Aleatoriedad:** Cada árbol se entrena con una muestra aleatoria de los datos de entrenamiento (con reemplazo, conocido como *bagging* o *bootstrap aggregation*). Además, en cada división de un nodo en el árbol, solo se considera un subconjunto aleatorio de características (variables) para encontrar la mejor división.
3.  **Votación/Promedio:**
    *   **Para clasificación:** Cuando se realiza una predicción, cada árbol 'vota' por una clase, y la clase con la mayoría de votos es la predicción final del modelo.
    *   **Para regresión:** La predicción final es el promedio de las predicciones de todos los árboles individuales.

**Ventajas clave:**

*   **Reduce el sobreajuste:** Al combinar muchos árboles, la varianza del modelo se reduce significativamente en comparación con un solo árbol de decisión, lo que ayuda a evitar el sobreajuste.
*   **Alta precisión:** Generalmente ofrece una alta precisión en una amplia gama de problemas.
*   **Maneja datos complejos:** Puede manejar datos con muchas características, desequilibrio de clases y valores atípicos.
*   **Selección de características:** Puede proporcionar una estimación de la importancia de cada característica.

````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, matthews_corrcoef, classification_report, confusion_matrix
````

````python
from sklearn.datasets import load_wine

# Cargar el dataset Wine
wine = load_wine()
X = wine.data  # Características
y = wine.target # Etiquetas de clase

# Convertir a DataFrame para una mejor visualización de las características
df_wine = pd.DataFrame(X, columns=wine.feature_names)
df_wine['target'] = y

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y) # Usamos stratify para mantener la proporción de clases

print(X_train.shape)

model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
model.fit(X_train, y_train)

print("\nModelo Random Forest entrenado exitosamente.")


# Realizar predicciones en el conjunto de prueba
y_pred = model.predict(X_test)

# Evaluar el rendimiento del modelo
mcc = matthews_corrcoef(y_test, y_pred)
print(f"MCC del modelo Random Forest: {mcc:.4f}")

print("\nReporte de Clasificación:\n")
print(classification_report(y_test, y_pred, target_names=wine.target_names))

print("\nMatriz de Confusión:\n")
print(confusion_matrix(y_test, y_pred))

# También podemos obtener la importancia de las características
feature_importances = model.feature_importances_
features = wine.feature_names
importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

print("\nImportancia de las características:\n")
display(importance_df)

````
