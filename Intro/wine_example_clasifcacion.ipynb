### ¿Qué es un Problema de Clasificación?

En el aprendizaje automático (Machine Learning), la **clasificación** es una tarea de aprendizaje supervisado que consiste en predecir la etiqueta o categoría a la que pertenece una instancia de datos.

**Elementos clave:**
1. **Características (Features):** Son las variables de entrada (en tu código, los componentes químicos del vino como alcohol, magnesio, etc.).
2. **Etiquetas (Labels/Targets):** Son las categorías de salida que queremos predecir (en este caso, el tipo de uva o cultivador: 0, 1 o 2).
3. **Modelo:** El algoritmo aprende patrones en las características para asignar la etiqueta correcta a nuevos datos.

**Ejemplo en este Notebook:**
El gráfico de **t-SNE** que has generado muestra cómo los datos se agrupan en el espacio. Cada color representa una 'clase'. Un buen modelo de clasificación sería capaz de dibujar fronteras entre estos grupos para separar correctamente los tres tipos de vino.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# 1. Cargar datos
wine = load_wine()
X_scaled = StandardScaler().fit_transform(wine.data)
y = wine.target # Las 3 clases de vino


# 2. t-SNE a 2 Dimensiones (para ver clusters)
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

# 3. Graficar
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', s=60, edgecolors='k')

# Añadir leyenda de clases
plt.legend(handles=scatter.legend_elements()[0], labels=['Uva 0', 'Uva 1', 'Uva 2'])

plt.title('Clasificación', fontsize=14)
plt.xlabel('Dimensión t-SNE 1', fontsize=12)
plt.ylabel('Dimensión t-SNE 2', fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
```
