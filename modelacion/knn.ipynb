### ¿Qué es KNN (K-Nearest Neighbors)?

**KNN** (K-Nearest Neighbors o K-Vecinos Más Cercanos) es un algoritmo de aprendizaje supervisado no paramétrico y muy sencillo que puede utilizarse tanto para tareas de clasificación como de regresión.

La idea principal de KNN es la siguiente:

1.  **Clasificación:** Para clasificar un nuevo punto de datos, el algoritmo encuentra los 'K' puntos de datos más cercanos (vecinos) en el conjunto de entrenamiento. La clase del nuevo punto se asigna entonces por la mayoría de las clases entre esos 'K' vecinos.
2.  **Regresión:** Para predecir el valor de un nuevo punto, se toma el promedio de los valores de los 'K' vecinos más cercanos.

La 'cercanía' entre puntos se calcula típicamente utilizando métricas de distancia como la distancia euclidiana, la distancia de Manhattan o la distancia de Minkowski.

### Implementación en Python

```python
from sklearn.neighbors import KNeighborsClassifier

```

````python
# Importar librerías necesarias
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris


iris = load_iris()

# 2. Extraer las características (X) y las etiquetas objetivo (y)
X = iris.data
y = iris.target

print(f"Dimensiones de X: {X.shape}")
print(f"Dimensiones de y: {y.shape}")


# 2. Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


# 3. Inicializar el modelo KNN
# Se elige k=5, lo que significa que el algoritmo considerará los 5 vecinos más cercanos
knn = KNeighborsClassifier(n_neighbors=5)

# 4. Entrenar el modelo
knn.fit(X_train, y_train)

# 5. Hacer predicciones sobre el conjunto de prueba
y_pred = knn.predict(X_test)

# 6. Evaluar el modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Precisión del modelo KNN con k=5: {accuracy:.2f}")

# Visualizar las predicciones
plt.figure(figsize=(8, 6))
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred, cmap='viridis', s=50, alpha=0.8, edgecolor='k')
plt.title('Predicciones del Modelo KNN')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.show()
````
