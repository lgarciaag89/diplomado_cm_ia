### ¿Qué es la Información Mutua (MI)?

La **Información Mutua** es una medida estadística que cuantifica cuánta información se comparte entre dos variables aleatorias. En términos sencillos, responde a la pregunta: *"Si conozco el valor de X, ¿cuánto reduce esto mi incertidumbre sobre el valor de Y?"*

**Características clave:**
*   **Relaciones No Lineales:** A diferencia de la correlación de Pearson (que solo detecta relaciones lineales), la MI puede capturar cualquier tipo de dependencia, incluyendo curvas y estructuras complejas.
*   **Simetría:** La información que X nos da sobre Y es la misma que Y nos da sobre X ($I(X;Y) = I(Y;X)$).
*   **No Negatividad:** Siempre es mayor o igual a cero. Un valor de 0 indica que las variables son completamente independientes.
*   **Interpretación en Bits:** Cuando se usa el logaritmo en base 2, se mide en 'bits', lo que permite una analogía directa con la teoría de conjuntos y los diagramas de Venn.

### Definición Matemática y Casos Extremos

Formalmente, para dos variables aleatorias discretas $X$ e $Y$, la **Información Mutua** se define como:

$$I(X;Y) = H(Y)-H(Y|X) = \sum_{y \in Y} \sum_{x \in X} p(x,y) \log\left(\frac{p(x,y)}{p(x)p(y)}\right)$$

Donde $p(x,y)$ es la función de probabilidad conjunta, y $p(x)$, $p(y)$ son las funciones de probabilidad marginal.

#### Casos Extremos:

1. **Independencia Total ($I(X;Y) = 0$):**
   - Ocurre cuando $p(x,y) = p(x)p(y)$.
   - Conocer el valor de $X$ no aporta absolutamente ninguna pista sobre el valor de $Y$. Visualmente, los círculos en un diagrama de Venn no se tocan.

2. **Dependencia Máxima / Determinista ($I(X;Y) = H(X) = H(Y)$):**
   - Ocurre cuando una variable es una función determinista de la otra (por ejemplo, $Y = X$).
   - Conocer $X$ elimina toda la incertidumbre sobre $Y$. En este caso, la Información Mutua es igual a la Entropía de las variables, y los círculos en el diagrama de Venn se solapan por completo.

### Informacion mutua para X1

Tabla observada:

| \($X_1$\) | \(Y\) |
|----------|------|
| 3 | 1 |
| 4 | 2 |
| 3 | 1 |
| 4 | 2 |

Probabilidades condicionales:

$$
P(Y=1 \mid X_1=3)=1
$$

$$
P(Y=2 \mid X_1=4)=1
$$

Entropia Condicional:

$$
H(Y \mid X_1) = -\Big(
\begin{aligned}
& P(X_1=3)*\big[P(Y=1 \mid X_1=3)\log P(Y=1 \mid X_1=3)\big] \\
& + P(X_1=4)*\big[
P(Y=1 \mid X_1=4)\log P(Y=1 \mid X_1=4) \big]
\end{aligned}
\Big)
$$

$$
\begin{aligned}
\\
H(Y \mid X_1) = P(X_1=3)*0 + P(X_1=4)*0 = 0
\end{aligned}
$$

Entonces:

$$
MI(X_1;Y)=H(Y)-H(Y \mid X_1) = 0.693-0 = 0.693
$$


**Como**:
$$
\begin{aligned}
\\
H(Y \mid X_1) = 0
\end{aligned}
$$

esto significa que:
- Si conoces $X_1$, sabes exactamente qué valor toma $Y$.
- No hay incertidumbre.

**Entonces**:
$$
\begin{aligned}
\\
MI = H(Y)
\end{aligned}
$$

- $X_1$ contiene toda la información posible sobre $Y$.
- Es el máximo que puede tomar la MI.

````python
from scipy.stats import pearsonr
def calcular_correlacion(x, y):
    """
    Calcula la correlación de Pearson.
    """
    corr, _ = pearsonr(x, y)
    return corr
````

````python
import numpy as np
def entropy(x):
  valores, counts = np.unique(x, return_counts=True)
  probs = counts / counts.sum()
  # Calculate Shannon entropy: -sum(p * log(p))
  return -(probs * np.log(probs)).sum()
````

````python

from sklearn.feature_selection import mutual_info_classif
def calcular_informacion_mutua_clas(x, y):
    """
    Calcula la Información Mutua usando mutual_info_regression.
    """
    mi = mutual_info_classif(x.reshape(-1, 1),y, discrete_features=True)[0]

    return entropy(x), entropy(y), mi
````

````python

from scipy.stats import pearsonr
from sklearn.feature_selection import mutual_info_regression
def calcular_informacion_mutua_regresion(x, y):
    """
    Calcula la Información Mutua usando mutual_info_regression.
    """
    mi = mutual_info_regression(x.reshape(-1, 1),y)[0]

    return entropy(x), entropy(y), mi

````

````python
import seaborn as sns
import numpy as np
def plot_correlacion(ax, x, y, titulo):
    corr = calcular_correlacion(x, y)

    sns.scatterplot(x=x, y=y,
                    ax=ax,
                    color='teal',
                    alpha=0.6)

    m, b = np.polyfit(x, y, 1)
    sns.lineplot(x=x, y=m*x + b,
                 ax=ax,
                 color='red')

    ax.set_title(f"{titulo} — Pearson r = {corr:.3f}",
                 fontsize=12,
                 fontweight='bold')

    ax.grid(True, linestyle='--', alpha=0.5)
````

````python
from matplotlib_venn import venn2
def plot_informacion_mutua(ax, x, y, titulo, type='clas'):
    """
    Crea el diagrama de Venn usando:
    - Entropía estimada
    - MI estimada por sklearn
    """

    if type == "clas":
        Hx, Hy, MI = calcular_informacion_mutua_clas(x, y)
    else:
      Hx, Hy, MI = calcular_informacion_mutua_regresion(x, y)
    # Redondeo
    Hx_r = round(Hx, 3)
    Hy_r = round(Hy, 3)
    MI_r = round(MI, 3)
    print(f"Hx: {Hx_r}\nHy: {Hy_r}\nMI: {MI_r}")

    v = venn2(
        subsets=(max(Hx_r - MI_r, 0),
                 max(Hy_r - MI_r, 0),
                 MI_r),
        set_labels=("H(X)", "H(Y)"),
        ax=ax
    )

    # Ajustar textos
    if v.get_label_by_id('10'):
        v.get_label_by_id('10').set_text(f"{max(Hx_r - MI_r,0):.3f}")
    if v.get_label_by_id('01'):
        v.get_label_by_id('01').set_text(f"{max(Hy_r - MI_r,0):.3f}")
    if v.get_label_by_id('11'):
        v.get_label_by_id('11').set_text(f"{MI_r:.3f}")

    ax.set_title(f"{titulo} — I(X;Y) = {MI_r:.3f}",
                 fontsize=12,
                 fontweight='bold')
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib_venn import venn2
from sklearn.preprocessing import LabelEncoder

# 1. Definir universo de categorías
categorias = ['A', 'B', 'C', 'D', 'E']
np.random.seed(42)
n = 500

# Generar X
x_disc = np.random.choice(categorias, size=n)

# Escenario 1: Dependencia Total (Y = X)
y_total_disc = x_disc.copy()

# Escenario 2: Dependencia Parcial (Y es X con 30% de ruido)
y_partial_disc = x_disc.copy()
mask = np.random.random(n) < 0.3
y_partial_disc[mask] = np.random.choice(categorias, size=mask.sum())

# Escenario 3: Independencia Total
y_indep_disc = np.random.choice(categorias, size=n)

print("Datos discretos generados exitosamente.")

import matplotlib.pyplot as plt

scenarios_disc = [
    ("Independiente", x_disc, y_indep_disc),
     ("Parcial (70% ident)", x_disc, y_partial_disc),
    ("Total (Y=X)", x_disc, y_total_disc)
]

fig, axes = plt.subplots(3, 2, figsize=(14, 20))
sns.set_theme(style="white")

for i, (titulo, x, y) in enumerate(scenarios_disc):
    le_x = LabelEncoder()
    le_y = LabelEncoder()
    x_enc = le_x.fit_transform(x)
    y_enc = le_y.fit_transform(y)
    # En datos discretos, nuestra función entropy() y calcular_informacion_mutua()
    # (usando mutual_info_regression sobre labels numéricos) funcionarán
    # de forma más alineada a la teoría si discretizamos correctamente.
    plot_correlacion(axes[i, 0], x_enc, y_enc, titulo)
    plot_informacion_mutua(axes[i,1], x_enc, y_enc, titulo,'clas')

plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib_venn import venn2

np.random.seed(42)
n = 500
x = np.linspace(-1, 1, n)

y_linear = x + np.random.normal(0, 0.1, n)
y_quad = x**2 + np.random.normal(0, 0.05, n)
y_noise = np.random.normal(0, 1, n)

scenarios = [
        ("Ruido", x, y_noise),
        ("Cuadrática", x, y_quad),
        ("Lineal", x, y_linear),
        ("Equals", x, x),
    ]

# Corregido de (3, 2) a (4, 2) para incluir todos los escenarios
fig, axes = plt.subplots(4, 2, figsize=(14, 20))
sns.set_theme(style="white")

for i, (titulo, data_x, data_y) in enumerate(scenarios):
  plot_correlacion(axes[i, 0], data_x, data_y, titulo)
  plot_informacion_mutua(axes[i, 1], data_x, data_y, titulo,'reg')

plt.tight_layout()
plt.show()
````
